## Construção de um Pipeline de dados

### Descrição
O projeto foi desenvolvido para consolidar duas base dados distintas de unidades diferentes empresas de uma corporação. O projeto irá construir uma estrutura de dados para tratar de forma recorrente essa bases de dados.



### Preparação do ambiente de trabalho

#### Instalação do WSL e do linux no ambinete Windows

- Instale o Windows Terminal na maquina caso ainda não esteja instalado. 
    - O Terminal do Windows trata-se de um aplicativo que permite utilizar diferentes linhas de comando como Prompt de Comando, PowerShell e o WSL.

- Execute o Terminal como administrador.
    - O terminal deverá abrir no Power Shell, Caso isso não ocorra, utilizando a barra de opções deve-se selecionar o Power Shell.

- No Power Shell execute o comando 

    - **WSL --list --online**

- Verifique na relação apresentada se existe a versão Ubuntu-22.04, e execute a linah de código.

    - **wsl --install -d Ubuntu-22.04**

- Após a instalação, utilize a barra de opões para selecionar o Ubuntu 22.04, para abrir o Ubuntu.


#### Instalação de pacotes

Antes de iniciar a instalação do **Python** será necessário atualizar os pacotes do **WSL**.

- Para atualizar os pacotes utilize os comandos.

    - **sudo apt-get update**

    - **sudo apt-get upgrade -y**

- Após a atualização dos pacotes do **WSL** erifique a versão do **Python** instalada com o comando.
    
    - **python3 --version** ou **python -V**

- Se o python não estiver instaldado utilize o comando a seguir para instalar

    - **sudo apt install python3.10**

- Após instalação do python, utilize os comando abaixo para instalação das bibliotecas **pip** e **venv**

    - **sudo apt install python3-pip -y**

    - **sudo apt install python3-venv -y**


#### Criação da estrutura de armazenamento do projeto

- Com linux aberto utilize o comando **ls** para mostrar todas a pastas existentes no diretório.

- Utilize o comando **mkdir** para cria uma nova pasta.

- Crie uma pasta chamada 'Documento'

- Utilize o comnado **cd Documentos** para entrar dentro da pasta documentos

- Dentro da pasta Documentos crie a seguinte estrutura

    - Documentos
        - pipeline_dados
            - data_raw
            - data_processed
            - notebooks
            - scripts

- Pastas:
    - pipeline_dados: Pasta do projeto
    - data_raw: Pasta destinada ao armazenamanto dos dados brutos
    - data_Processed: Pasta destinada ao aramenamento dos dados processados
    - notebooks: Pasta destinada ao armazenamento dos notebooks criados
    - scripts: pasta destinada ao armazenamento dos scripts de python 


#### Criação de ambiente virtual 

Antes de criarmos o ambiente virtual em **python**, precisamos acessar a pasta do projeto **pipeline_dados**

- Para criarmos o amabiente virtual em **python** utilizamos o comandos a seguir

    - **python3 -m venv venv**

- O ambiente criado deve ser ativado com o comando a seguir.

    - **source venv/bin/activate**

#### Instalando bibliotecas

Neste projeto iremos utiliar as biblioteca **Pandas** e **Requests**.

 - Para instalar as pandas e requests podemos utilizar os comandos 

    - **pip install requests==2.28.2** 

    - **pip install pandas==2.0.0**

#### Instalação  e configuração do VS Code

Para instalar o VS Code basta ir no site offical da ferramente e realizar o download.

- Durante a instalação deve-se selecionar a opção **Adicione em PATH (disponível após reiniciar)** na ciaxa de dialogo.

- Para utilizar o VS Code acesse a pasta **pipeline_dados** no **WSL** e utilize o comando a seguir
    
    - **Code .**

- Com o VS Code aberto instale as extensões 

    - **WSL**
    - **Python IntelliSense (Pylance)**

Está serão as configurações necessárias para a execução deste projeto.
